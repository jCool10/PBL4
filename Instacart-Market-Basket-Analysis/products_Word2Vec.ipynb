```python
import pandas as pd
import numpy as np
import gensim
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
train_orders = pd.read_csv("input/order_products__train.csv")
# prior_orders = pd.read_csv("input/order_products__prior.csv")
products = pd.read_csv("input/products.csv").set_index('product_id')
```

```python
#Turn the product ID to a string
#This is necessary because Gensim's Word2Vec expects sentences, so we have to resort to this dirty workaround
train_orders["product_id"] = train_orders["product_id"].astype(str)
# prior_orders["product_id"] = prior_orders["product_id"].astype(str)
```

```python
train_products = train_orders.groupby("order_id").apply(lambda order: order['product_id'].tolist()).tolist()
# prior_products = prior_orders.groupby("order_id").apply(lambda order: order['product_id'].tolist()).tolist()
```

```python
sentences = train_products
sentences
```

```python
model = gensim.models.Word2Vec(sentences, vector_size=100, window=5, min_count=5, workers=4)

print(model)
```

```python

model.save("features/product2vec.model")
model.wv.save_word2vec_format("features/product2vec.model.bin", binary=True)
```

```python
model = gensim.models.Word2Vec.load("features/product2vec.model")

model.wv.key_to_index
```

```python
embedding_matrix = np.zeros((49688, 100))
for i in range(1,49689):
    word=str(i)
    if word in model.wv.key_to_index:
        embedding_matrix[i-1] = model.wv[word]
```

```python
from sklearn.decomposition import PCA

features_num=2
pca=PCA(features_num)
embedding_matrix=pca.fit_transform(embedding_matrix)
features=pd.DataFrame(embedding_matrix)
features.columns=["w2v_%s"%j for j in range(features_num)]
features["product_id"]=range(1,49689)
print (features.head())

features.to_csv("features/product2vec.csv",index=None)
```
